```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imageFile = 'Lena.png'

# 1. BGR
imgBGR = cv2.imread(imageFile)
if imgBGR is None:
    print("cannot find " , imageFile)
    exit()
plt.axis('off')
plt.imshow(imgBGR[:,:,::-1])
plt.title('BGR: mismatch of channels')
plt.show()
```

```python
# 2. RGB
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.title('RGB by cvtColor')
plt.show()
```

```python
imgHSV = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2HSV)
print(imgHSV.shape, imgHSV.dtype)
```

```python
plt.subplot(1,3,1)
plt.imshow(imgHSV[:,:,0], cmap = 'gray')  #H
plt.subplot(1,3,2)
plt.imshow(imgHSV[:,:,1], cmap = 'gray')  #S
plt.subplot(1,3,3)
plt.imshow(imgHSV[:,:,2], cmap = 'gray')  #V
```
