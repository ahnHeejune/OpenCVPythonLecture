```python
# 패키지 임포트 
import numpy as np
import cv2 # as cv
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lena.png")
if img is None:
    print("Cannot load the file")    
else:
    print(type(img))
    print(img.dtype, img.shape)
```

```python
print(f"type:{type(img)}")
print(f"shape:{img.shape}, dtype:{img.dtype}")
```

```python
plt.subplot(1,3,1)
plt.imshow(img)
#plt.axis('off')
plt.title('BGR')
plt.subplot(1,3,2)
plt.imshow(img[:,:,::-1])  # BGR => RGB
plt.axis('off')
plt.title('BGR[::-1]')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(1,3,3)
plt.imshow(img2)
plt.axis('on')
plt.title('cvtColor')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
print(f"shape:{img_gray.shape}, dtype:{img_gray.dtype}")
```

```python
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

```python
b1, g1, r1 = img[:,:,0], img[:,:,1], img[:,:,2]
print(b1.shape)
b, g, r= cv2.split(img)
print(b.shape)
#imgRGB2 = cv2.merge((r,g,b))
#imgBGR2 = cv2.merge((b,g,r))
```

```python
print(f"shape:{b1.shape}, dtype:{b1.dtype}")
print(f"shape:{b.shape}, dtype:{b.dtype}")
```

```python
plt.subplot(2,3,1)
plt.imshow(b1, cmap = 'gray')
plt.subplot(2,3,2)
plt.imshow(g1, cmap = 'gray')
plt.subplot(2,3,3)
plt.imshow(r1, cmap = 'gray')

plt.subplot(2,3,4)
plt.imshow(b, cmap = 'gray')
plt.subplot(2,3,5)
plt.imshow(g, cmap = 'gray')
plt.subplot(2,3,6)
plt.imshow(r, cmap = 'gray')
plt.show()
```

```python
imgBGR2 = cv2.merge((b,g,r))
imgRGB2 = cv2.merge((r,g,b))
plt.subplot(1,2,1)
plt.imshow(imgBGR2)
plt.subplot(1,2,2)
plt.imshow(imgRGB2)
plt.show()
```
